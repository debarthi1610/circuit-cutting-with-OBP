```python
import os
from qiskit import QuantumCircuit

qasm_dir = "/Users/debarthipal/Library/CloudStorage/OneDrive-IBM/Desktop/VS files/Circuit cutting+OBP/QASM_circuits_large"
qasm_files = sorted([
    f for f in os.listdir(qasm_dir)
    if f.endswith(".qasm")
])
print(qasm_files)
```

```python
from circuit_cutting import do_cc
from qiskit_ibm_runtime.fake_provider import FakeTorino
backend = FakeTorino()
```

```python
import json
circ_dic={}
for index, file in enumerate(qasm_files, start=1):

    n=QuantumCircuit.from_qasm_file(file).num_qubits
    
    circuit=QuantumCircuit.from_qasm_file(file)
    number_of_cuts, sampling_overhead_values, subcirc_length = do_cc(circuit, backend)
    circ_dic[index] = {"filename":file, 
                          "num_cuts": number_of_cuts, 
                          "sampling_overhead":sampling_overhead_values,
                          "subcirc_length":subcirc_length
                          }
    
    json_obj=json.dumps(circ_dic, indent=4)
    with open("cuts_benchpress_large.json", "w") as f:
        f.write(json_obj)   #Saving to a JSON file
```
